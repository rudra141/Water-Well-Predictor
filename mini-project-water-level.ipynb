```python
def detect_outliers_iqr(data):
  outlier_list = []
  data = sorted(data)
  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)
  #print("The Val of Q1 and Q2",q1, q3)
  IQR = q3-q1
  lwr_bound = q1-(1.5*IQR)
  upr_bound = q3+(1.5*IQR)
  #print("The lower & Upper Bound",lwr_bound, upr_bound)
  for i in data:
    if (i<lwr_bound or i>upr_bound):
      outlier_list.append(i)
  return outlier_list
```

```python
for i in features:
  outliers = detect_outliers_iqr(filtered_df[i])
  print("Outliers in",i,"attribute :", outliers)
```

```python
def handle_outliers(data):
  tenth_percentile = np.percentile(data, 25)
  ninetieth_percentile = np.percentile(data, 75)
  b = np.where(data<tenth_percentile, tenth_percentile, data)
  b1 = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
  return b1

for i in features:
  filtered_df[i]=handle_outliers(filtered_df[i])
```

```python
import pandas as pd

months = ['Jan', 'May', 'Aug', 'Nov']

mean_df = pd.DataFrame()

for month in months:
    columns = [col for col in filtered_df.columns if month in col]
    month_data = filtered_df[columns]
    
    # Exclude rows where all values are NaN
    non_nan_rows = month_data.notna().any(axis=1)
    month_data = month_data[non_nan_rows]

    # Calculate mean for each row
    mean_values = month_data.mean(axis=1)
    
    mean_df[month] = mean_values

# Now mean_df should not contain any NaN values
```
